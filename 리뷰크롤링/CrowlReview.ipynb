```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

cafe_urls = [
    "https://map.naver.com/p/search/건대입구%20카페/place/1982514659?c=17.22,0,0,0,dh&placePath=/review",
    "https://map.naver.com/p/search/건대입구%20카페/place/1872641035?c=17.22,0,0,0,dh&placePath=/review"
]

all_reviews = []

for url in cafe_urls:
    print(f"▶ 크롤링 중: {url}")
    driver.get(url)
    time.sleep(5)

    # 주소 추출을 위해 /home URL로 이동
    home_url = url.replace('/review', '/home')
    driver.get(home_url)
    time.sleep(3)
    try:
        driver.switch_to.default_content()
        driver.switch_to.frame("entryIframe")
        address = driver.find_element(By.CSS_SELECTOR, 'span.LDgIH').text.strip()
    except:
        address = "주소_추출실패"

    # 다시 리뷰 페이지로 이동
    driver.get(url)
    time.sleep(5)
    try:
        driver.switch_to.default_content()
        driver.switch_to.frame("entryIframe")
    except:
        print("❗ iframe 진입 실패:", url)
        continue

    time.sleep(2)

    # 카페 이름 추출
    try:
        cafe_name = driver.find_element(By.CSS_SELECTOR, 'span.GHAhO').text.strip()
    except:
        cafe_name = "이름_추출실패"

    for _ in range(15):
        driver.execute_script("window.scrollBy(0, 1000);")
        time.sleep(1)

    reviews = driver.find_elements(
        By.CSS_SELECTOR,
        'a[role="button"][data-pui-click-code="rvshowmore"], a[role="button"][data-pui-click-code="rvshowless"]'
    )

    for r in reviews:
        text = r.get_attribute("innerText").strip().replace("\n", " ")
        if text and len(text) > 10:
            all_reviews.append({
                '카페이름': cafe_name,
                '주소': address,
                '카페URL': url,
                '리뷰': text
            })

driver.quit()

# 컬럼 순서 지정
df = pd.DataFrame(all_reviews, columns=['카페이름', '주소', '카페URL', '리뷰'])
df.to_csv("건대입구_10개카페_리뷰.csv", index=False, encoding='utf-8-sig')
print("✅ 전체 수집된 리뷰 수:", len(df))
```
